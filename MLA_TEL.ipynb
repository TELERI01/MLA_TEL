```python
import pandas as pd

# Replace 'your_file_path' with the path to your file in Google Drive
file_path = '/content/drive/MyDrive/Bellarmine/CS430/MLA_TEL/weather_2012.csv'

# Read the file into a DataFrame
df = pd.read_csv(file_path)

# Display the first few rows of the DataFrame
df.head()
```

```python
print("\nLast five rows:")
print(df.tail())
```

```python
print("\nRandom sample of five rows:")
print(df.sample(5))
```

```python
# Step 2: Display a list of all the columns in the dataset
print("\nColumns in the dataset:")
print(df.columns)
```

```python
# Step 3: Rename the columns
df.columns = ['date_time', 'temp_c', 'dew_point', 'relative_humidity', 'wind_speed', 'visibility', 'pressure', 'weather']
print("\nRenamed columns:")
print(df.columns)
```

```python
# Step 4: Get standard statistics for each column
print("\nStandard statistics for each column:")
print(df.describe())
```

```python
# Step 5: Add a new column 'temp_f' that converts 'temp_c' to Fahrenheit
df['temp_f'] = df['temp_c'] * 9/5 + 32
```

```python
# Step 6: Rearrange the columns so 'temp_f' appears immediately before 'temp_c'
cols = df.columns.tolist()
cols.insert(cols.index('temp_c'), cols.pop(cols.index('temp_f')))
df = df[cols]

print("\nColumns after rearrangement:")
print(df.head())
```

```python
# Step 7: Display the data types of all the columns
print("\nData types of all columns:")
print(df.dtypes)
```

```python
# Step 8: Convert the 'date_time' column to datetime and verify the conversion
df['date_time'] = pd.to_datetime(df['date_time'])
print("\nVerifying conversion to datetime:")
print(df['date_time'].dtypes)
```

```python
# Step 9: Get the average temperature (in Fahrenheit) for each month and store it in a Pandas Series
df['month'] = df['date_time'].dt.month
monthly_avg = df.groupby('month')['temp_f'].mean()

print("\nMonthly average temperature (F):")
print(monthly_avg)
```

```python
import matplotlib.pyplot as plt

# Step 10: Generate a scatterplot that compares temperature (in F) against relative humidity
plt.figure(figsize=(20, 10))
plt.scatter(df['temp_f'], df['relative_humidity'])
plt.title('Temperature vs Relative Humidity')
plt.xlabel('Temperature (F)')
plt.ylabel('Relative Humidity')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Step 11: Generate a line chart that shows the average temperature per month
plt.figure(figsize=(20, 10))
plt.plot(monthly_avg.index, monthly_avg.values, marker='o', linestyle='-', color='b')

# Add labels and title
plt.title('Average Temperature per Month (in Fahrenheit)', fontsize=16)
plt.xlabel('Month', fontsize=14)
plt.ylabel('Average Temperature (F)', fontsize=14)

# Display the plot
plt.grid(True)
plt.xticks(monthly_avg.index)  # Ensure each month is labeled
plt.show()
```
